```python
from abc import ABC, abstractmethod

class Figura(ABC): # Base (Padre)
    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimetro(self):
        pass

    def mostrar(self):
        print("Esto es un figura geométrica.")
```

```python
class Circulo(Figura): # Derivada (Hija)
    def __init__(self, radio):
        self.radio = radio

    def area(self):
        return 3.1416 * self.radio ** 2

    def perimetro(self):
        return 2 * 3.1416 * self.radio
    
    def mostrar(self):
        print(f"Radio: {self.radio}")
```

```python
objeto_circulo = Circulo(5)
objeto_circulo.mostrar()
print(f"Área del círculo: {objeto_circulo.area()}")
print(f"Perímetro del círculo: {objeto_circulo.perimetro()}")
```

# Ejercicio 1

```python
class Accionar:
    def encender(self):
        print("Encendiendo...")
    def apagar(self):
        print("Apagando...")
```

```python
from abc import ABC, abstractmethod

class Electrodomestico(ABC):
    def __init__(self, potenciaKW, marca):
        self.potenciaKW = potenciaKW
        self.marca = marca
    
    @abstractmethod
    def consumo_energetico(self, horas):
        pass

    def __str__(self):
        return f"Electrodoméstico: {self.marca}, Potencia: {self.potenciaKW} kW"
    
    def encender(self,objeto_accionar):
        if isinstance(objeto_accionar, Accionar):
            objeto_accionar.encender()
        else:
            print("El objeto no tiene la capacidad de encender.")
    
    def apagar(self,objeto_accionar):
        if isinstance(objeto_accionar, Accionar):
            objeto_accionar.apagar()
        else:
            print("El objeto no tiene la capacidad de apagar.")
```

```python
class Coccion(Electrodomestico):

    def __init__(self, potenciaKW, marca, programable):
        super().__init__(potenciaKW, marca)
        self.programable = programable

    def consumo_energetico(self, costo_kwh):
        return self.potenciaKW * costo_kwh

    def __str__(self):
        return f"Cocción: {self.marca}, Potencia: {self.potenciaKW} kW, Programable: {self.programable}"
```

```python
class Cocina (Coccion):
    def __init__(self, potenciaKW, marca, programable, hornillas):
        super().__init__(potenciaKW, marca, programable)
        self.hornillas = hornillas
    
    def consumo_energetico(self, costo_kwh):
        return self.potenciaKW * costo_kwh * self.hornillas

    def __str__(self):
        return f"Cocina: {self.marca}, Potencia: {self.potenciaKW} kW, Programable: {self.programable}, Hornillas: {self.hornillas}"
```

```python
class Horno (Coccion):
    def __init__(self, potenciaKW, marca, programable, capacidad):
        super().__init__(potenciaKW, marca, programable)
        self.capacidad = capacidad

    def consumo_energetico(self, costo_kwh):
        return self.potenciaKW * costo_kwh * self.capacidad

    def __str__(self):
        return f"Horno: {self.marca}, Potencia: {self.potenciaKW} kW, Programable: {self.programable}, Capacidad: {self.capacidad}"
```

```python
# Crear objetos de Accionar
objeto_accionar = Accionar()

# Craer objetos de Cocina y Horno
objeto_cocina = Cocina(2.5,"Samsung", True, 4)
objeto_horno = Horno(3.0, "LG", True,50)

#Mostrar la información de los objetos
print(objeto_cocina)
objeto_cocina.encender(objeto_accionar)
objeto_cocina.apagar(objeto_accionar)
print(f'Consumo energético de la cocina: {objeto_cocina.consumo_energetico(costo_kwh=0.15)}')

print(objeto_horno)
objeto_horno.encender(objeto_accionar)
objeto_horno.apagar(objeto_accionar)
print(f'Consumo energético del horno: {objeto_horno.consumo_energetico(costo_kwh=0.15)}')
```

```python
from abc import ABC, abstractmethod

class Volumen(ABC):
    def __init__(self, id,name):    
        self.id = id
        self.name = name
    
    @abstractmethod
    def __str__(self):
        pass

    @property
    def id(self):
        return self._id

    @id.setter
    def id(self, value):
        self._id = value

    @property
    def name(self):
        return self._name

    @name.setter
    def name(self, value):
        self._name = value
```

```python
class Book(Volumen):
    def __init__(self, id, name, isbn_code):
        super().__init__(id, name)
        self.isbn_code = isbn_code

    def __str__(self):
        return f"Book: {self.name}, ISBN: {self.isbn_code}"
```

```python
class NewPaper(Volumen):
    def __init__(self, id, name, issn_code):
        super().__init__(id, name)
        self.issn_code = issn_code

    def __str__(self):
        return f"Book: {self.name}, ISSN: {self.issn_code}"
```

```python
import pandas as pd
class Library:
    def __init__(self,name,address):
        self.name = name
        self.address = address
        self.list_volumenes = []
    
    def add_volumen(self, objeto_volumen):
        self.list_volumenes.append(objeto_volumen)
    
    def remove_volumen(self, id):
        for volumen in self.list_volumenes:
            if volumen.id == id:
                self.list_volumenes.remove(volumen)
                return f"Volumen con ID {id} eliminado." #break
            
        return f"Volumen con ID {id} no encontrado." # print(f"Volumen con ID {id} no encontrado.")
    
    def save_volumenes(self):
        data_to_save = []
        for volumen in self.list_volumenes:
            data_to_save.append({
                'id': volumen.id,
                'name': volumen.name
            })
        
        df = pd.DataFrame(data_to_save)
        df.to_excel('volumenes.xlsx', index=False)

    def mostrar_volumenes(self):
        for volumen in self.list_volumenes:
            print(volumen)
```

```python
objeto_library = Library("Biblioteca Central", "Calle Principal 123")

objeto_book1 = Book(1, "El Quijote", "1234567890")
objeto_book2 = Book(2, "Cien Años de Soledad", "0987654321")

objeto_new_paper1 = NewPaper(3, "El País", "1234-5678")
objeto_new_paper2 = NewPaper(4, "La Nación", "8765-4321")

objeto_library.add_volumen(objeto_book1)
objeto_library.add_volumen(objeto_book2)
objeto_library.add_volumen(objeto_new_paper1)
objeto_library.add_volumen(objeto_new_paper2)

objeto_library.mostrar_volumenes()
objeto_library.remove_volumen(2)
print("\nDespués de eliminar el volumen con ID 2:\n")
objeto_library.mostrar_volumenes()

#Guardar en excel
objeto_library.save_volumenes()
```
